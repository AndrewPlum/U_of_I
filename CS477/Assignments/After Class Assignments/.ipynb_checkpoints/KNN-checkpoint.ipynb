```python
import numpy as np
```

```python
# Iris data

from sklearn import datasets as ds

iris = ds.load_iris()
display(iris)
```

```python
X = iris.data
Y = iris.target

print(X.shape, Y.shape)
print(Y)
```

```python
# split data in to training and test

from sklearn.model_selection import train_test_split

#X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2) 
X_train, Y_train, X_test, Y_test = train_test_split(X, Y, test_size = 0.2) 
# Whatever the test size parameter is set to is the percentage of the data that is split into training

print(X_train.shape)
print(X_train,"\n")
print(Y_train.shape)
print(Y_train,"\n")
print(X_test.shape)
print(X_test,"\n")
print(Y_test.shape)
print(Y_test,"\n")
```

```python
def Euclidean_Distance(x1, x2):
    euclidean_distance = 0
    euclidean_distance = np.sqrt(np.sum((x2-x1)**2))
    return euclidean_distance
```
