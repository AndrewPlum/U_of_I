##### Task 2.1: Complete the following function to identify the K nearest neighbors of a given data sample. 
- return both the indices and Euclidean distances of the K nearest neighbors for a given query/new sample.

```python
def findKNgbs(X_train, x_query, K = 5):
    '''find K nearest neighbors for a given data sample in X_train
        
        input:
            X_train: training set
            x_query: new data sample/a query
            K: the number of neighbors
        return:
            indK: the indices of of the K nearest neighbors
            disK: the distances of the K nearest neighbors to x_new (x_query?)          
    '''
    #add your code here
    
    # First way to calculate dis
    
    #index=0
    dis = np.zeros((X_train.shape[0])) # numpy array with 0s
    for i, x in enumerate(X_train):
        dis[i] = np.sqrt(np.sum((x - x_query)**2)) #euclidean dis between x and x_new
        #print(dis[i])
        #index += 1
        #print(dis[i], index)
    #print(dis)
    
    # Second way to calculate dis
    
    #dis = np.sqrt(np.sum((X_train - x_query)**2, axis=1))
    #print(dis)
    
    # Calculate indK
    
    indK = np.argsort(dis, axis = 0) # find the k nearest neighbors using a sorting algorithm

    # Calculate disK
    
    i = 0 
    disK = np.zeros(dis.shape[0])
    for ele in indK:
        #print(dis[ele])
        disK[i] = dis[ele]
        i += 1
    #disK2 = np.sort(dis, axis = 0)    
    
    # Checks
    
    #print(indK[0:10])
    #print(disK[0:10])
    #print(disK2[0:10])
    
    # Return indices and distances of K nearest neighbors
    
    return indK[:K], disK[:K] # return the indices and distances for the K nearest neighbors

    
    
x = X_test[0]
print('Input data sample:\n', x)
k_inx, k_dis = findKNgbs(X_train, x, K=5)
print('\nK nearest neighbors of x in the training set:\n', k_inx, '\n\nTheir distances to the query:\n',k_dis)
```


##### Task 2.2: Predict the target value for new data samples.


```python
def predict(X_in, X_train, y_train, K = 5):
    '''predict the target values for input queries
        
        input:
            X_in: new data samples/queries. n*4. contains multiple data samples
            X_train: the feature vectors of training samples
            y_train: the target values of the training samples
            K: the number of neighbors

        return:
            y_pred: the predictions of the input queries
    '''

    #add your code here
    
      
    # First way
    
    #print("Method 1:\n")
    
    y_pred = np.zeros((X_in.shape[0]))
    #print(y_pred.shape)

    i = 0 
    for x_in in X_in:
        dis = np.sqrt(np.sum((X_train - x_in)**2, axis=1))
        #print(dis)
        nearest_ind = np.argsort(dis)[:K]
        nearest_dis = dis[nearest_ind]
        #nearest_dis = np.sort(dis, axis = 0)[:K]
        #print(nearest_ind)
        #print(nearest_dis)
    
        nearest_tgts = y_train[nearest_ind].astype(int)
        predicted_label = np.mean(nearest_tgts)
        #print(predicted_label)
        y_pred[i] = predicted_label
        i += 1
    
    # Return prediction    
    
    return y_pred
    
    
    
n = 10 # 3 test samples
K = 6 # k neighbors

X_in = X_test[0:n]
y_pred = predict(X_in, X_train, y_train, K)
print('Predictions:', y_pred)
print('True target values:', y_test[:n])
```

#### Task 3: Evaluate the KNN method using MSE and MAE. 10 points. <a id="task3">

- Implement the mean square error(MSE) and mean absolute error(MAE) functions.
- Calcualte the MSE and MAE of the knn on the test set

    - MSE: ((y_true[0]-y_pred[0])**2 + ...+(y_true[n-1]-y_pred[n-1])**2))/n where n is the number of samples
    - MAE: (|y_true[0]-y_pred[0]| + ...+|y_true[n-1]-y_pred[n-1]|)/n

```python
# calculate mean square error (MSE) between y_true and y_pred
def myMSE(y_true, y_pred):
    '''
        y_true: the true target values
        y_pred: predictions

        return: the MSE between y_true and y_pred
    '''    
    #add your code here
    
    #t = y_true - y_pred
    #print(t.shape)
    
    # MSE
    
    MSE = np.mean((y_true - y_pred)**2)
    
    return MSE
    

    
def myMAE(y_true, y_pred):
    '''
        y_true: the true target values
        y_pred: predictions

        return: the MAE between y_true and y_pred
    '''    
    #add your code here

    # MAE
    
    MAE = np.mean(np.abs(y_true - y_pred))
    
    return MAE


    
#print(y_test.shape, y_pred.shape)        
y_pred = predict(X_test, X_train, y_train, K=4)
mse = myMSE(y_test, y_pred)
mae = myMAE(y_test, y_pred)
print('The MSE on the test set is', round(mse, 2))
print('The MAE on the test set is', round(mae, 2))
```

Weighted Average KNN:

```python
def weighted_average_predict(X_in, X_train, y_train, K = 5):
    '''predict the target values for input queries
        
        input:
            X_in: new data samples/queries. n*4. contains multiple data samples
            X_train: the feature vectors of training samples
            y_train: the target values of the training samples
            K: the number of neighbors

        return:
            y_pred: the predictions of the input queries
    '''

    #add your code here
    
      
    # First way
    
    #print("Method 1:\n")
    
    y_pred = np.zeros((X_in.shape[0]))
    #print(y_pred.shape)

    i = 0 
    for x_in in X_in:
        dis = np.sqrt(np.sum((X_train - x_in)**2, axis=1))
        #print(dis)
        nearest_ind = np.argsort(dis)[:K]
        nearest_dis = dis[nearest_ind]
        inverse_nearest_dis = 1 / nearest_dis
        #nearest_dis = np.sort(dis, axis = 0)[:K]
        #print(nearest_ind)
        #print(nearest_dis)
    
        nearest_tgts = y_train[nearest_ind].astype(int)
        predicted_label = np.average(nearest_tgts, weights = inverse_nearest_dis)
        #print(predicted_label)
        y_pred[i] = predicted_label
        i += 1
    
    # Return prediction    
    
    return y_pred

print("Function defined")
```

```python
# Calculate MSES

mses= []
l = []
for i in range(30):
    #print(i + 1)
    l.append(i + 1)
    current_K = i + 1
    y_pred = weighted_average_predict(X_test, X_train, y_train, K = current_K)
    #print("Current number of nearest neigbors (k):", current_K)
    mse = myMSE(y_test, y_pred)
    mae = myMAE(y_test, y_pred)
    mses.append(mse) 
    #mses.append(round(mse, 2)) # rounds to hundredths place
    """
    print('The MSE on the test set is', round(mse, 2))
    print('The MAE on the test set is', round(mae, 2))
    print("")
print("MSES:")
print(mses)
"""

# Plot MSES

plt.plot(l, mses, label = "Mean Square Errors", marker="x")
plt.xlabel("K")
plt.ylabel("MSE")
plt.title("Number of Nearest Neigbors (k) vs Mean Square Error")
plt.legend()
plt.show()
```

Minkowski Distance KNN:

```python
def minkowski_dis_predict(X_in, X_train, y_train, K = 9, p = 16):
    '''predict the target values for input queries
        
        input:
            X_in: new data samples/queries. n*4. contains multiple data samples
            X_train: the feature vectors of training samples
            y_train: the target values of the training samples
            K: the number of neighbors

        return:
            y_pred: the predictions of the input queries
    '''

    #add your code here
    
      
    # First way
    
    #print("Method 1:\n")
    
    y_pred = np.zeros((X_in.shape[0]))
    #print(y_pred.shape)

    i = 0 
    for x_in in X_in:
        dis = np.sum(np.abs(X_train - x_in)**p, axis=1)**(1/p) # calculate minkowski distance
        #print(dis)
        nearest_ind = np.argsort(dis)[:K]
        nearest_dis = dis[nearest_ind]
        #inverse_nearest_dis = 1 / nearest_dis
        #nearest_dis = np.sort(dis, axis = 0)[:K]
        #print(nearest_ind)
        #print(nearest_dis)
    
        nearest_tgts = y_train[nearest_ind].astype(int)
        predicted_label = np.mean(nearest_tgts)
        #predicted_label = np.average(nearest_tgts, weights = inverse_nearest_dis)
        #print(predicted_label)
        y_pred[i] = predicted_label
        i += 1
    
    # Return prediction    
    
    return y_pred

print("Function defined")
```

```python
# Calculate MSES

best_mses_list = []
l = []
for i in range(30):
    #print(i + 1)
    l.append(i + 1)
    current_K = i + 1
    
    mse_list = []
    for j in range(30):
        current_p = j + 1
        y_pred = minkowski_dis_predict(X_test, X_train, y_train, K = current_K, p = current_p)
        
        mse_list.append(myMSE(y_test, y_pred))
    best_mse = min(mse_list)
    best_mses_list.append(best_mse)     
        
    #print("Current number of nearest neigbors (k):", current_K)
    
    #mses.append(round(mse, 2)) # rounds to hundredths place
    #print('The MSE on the test set is', round(mse, 2))
    #print('The MAE on the test set is', round(mae, 2))
    #print("")
#print("MSES:")
#print(mses)
    
# Plot MSES

plt.plot(l, best_mses_list, label = "Mean Square Errors", marker="x")
plt.xlabel("K")
plt.ylabel("MSE")
plt.title("Number of Nearest Neigbors (k) vs Mean Square Error Using Best P")
plt.legend()
plt.show()
```
