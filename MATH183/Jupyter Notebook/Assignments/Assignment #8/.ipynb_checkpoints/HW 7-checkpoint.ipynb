## Part 1: Numpy and plotting 

__Problem 1:__(3D plotting) Plot the function $f(x,y) = \cos(x*y)$, $(x,y) \in [-\pi,\pi] \times [-\pi,\pi]$ in three dimensional space.

```python
# import packages

import numpy as np
import matplotlib.pyplot as plt

# initialize variables

x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x,y)

# define function and assign data from it to varible Z

def f(x,y):
    return np.cos(X * Y)

Z = f(X, Y)

# plot in 3D

%matplotlib notebook

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='plasma')
```

__Problem 2:__(Animation) Plot the function $f(x,t) = e^{-x^2+t}$ for $x \in [-3,3]$ and $t \in [0,2]$.

```python
# import packages

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook

# initialize variables x and t

x = np.linspace(-3, 3, 100)
t = np.linspace(0, 2, 100)

# define function and assign data from it to varible Z

def f(x, t):
    return np.exp(-x**2 + t)

# create figure and axes and set the axes limits

fig, ax = plt.subplots()
ax.set_xlim(-3, 3)
ax.set_ylim(0, 10)
line, = ax.plot([], [])

# define animate function

def animate(frame):
    line.set_data(x, f(x, frame))
    ax.set_title(f"t = {frame:.2f}") # display value of t (float) as animation title

# show animation

animation = FuncAnimation(fig, animate, frames=t, interval=100, blit=True)
```
