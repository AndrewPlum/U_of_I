__Problem:__ Segment all $24$ coins from the dataset "data.coins()" of skimage and label them. NOTE: make sure the segmentation looks as nice as possible.

```python
# retrieve coins image

from skimage import data,io

coins = data.coins()
io.imshow(coins)
```

```python
# binary and otsu

from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt

otsu = threshold_otsu(coins)
binary = coins > otsu
print(otsu)
plt.subplot(1,2,1)
io.imshow(coins)
plt.subplot(1,2,2)
io.imshow(binary)
```

```python
# Closing
from skimage.morphology import square, closing

B = square(4)
new_closed = closing(binary, B) 
io.imshow(new_closed)
```
