__Problem:__ Segment all $24$ coins from the dataset "data.coins()" of skimage and label them. NOTE: make sure the segmentation looks as nice as possible.

```python
# Import packages

import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color, io
from skimage.filters import threshold_otsu
from skimage.exposure import equalize_adapthist
from skimage.morphology import square, disk, opening, closing, dilation, erosion
from skimage.morphology import remove_small_objects
from skimage.segmentation import clear_border
from skimage.measure import label
```

```python
# Coins image

coins = data.coins()
io.imshow(coins)
new_coins = equalize_adapthist(coins, clip_limit = 0.01, nbins=256)

plt.subplot(1,2,1)
io.imshow(coins)

plt.subplot(1,2,2)
io.imshow(new_coins)
```

```python
# Otsu and binary

otsu = threshold_otsu(new_coins)
print(otsu)
binary = new_coins > otsu
io.imshow(binary)
```

```python
# Close

B = square(4)
new_closed = closing(binary, B) 
io.imshow(new_closed)
```

```python
# Clear border

cleared = clear_border(new_closed)
io.imshow(cleared)
```

```python
# Remove small objects

removed = remove_small_objects(cleared)
io.imshow(removed)
```

```python
# Open

new_cleared = opening(cleared)
io.imshow(new_cleared)
```

```python
# Remove small objects

new_removed = remove_small_objects(new_cleared)
io.imshow(new_removed)
```

```python
# Close

B = square(6)
new_closed = closing(new_removed, B)
io.imshow(new_closed)
```

```python
# Label 

new_label = label(new_closed)
io.imshow(new_label)
```

```python
# Plot the result

plt.subplot(1,2,1)
io.imshow(coins)
plt.subplot(1,2,2)
io.imshow(new_label*coins, cmap='plasma')
```

__Problem:__ Segement the $10$ largest stars from the dataset "data.hubble_deep_field()" of skimage and label them.

```python
# Import packages

import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color, io
from skimage.filters import threshold_otsu
from skimage.exposure import equalize_adapthist
from skimage.morphology import square, disk, opening, closing, dilation, erosion
from skimage.morphology import remove_small_objects
from skimage.segmentation import clear_border
from skimage.measure import label
```

```python
# Hubble deep field image

hubble = data.hubble_deep_field()
io.imshow(hubble)
```

```python
# Equalize adapt this

new_hubble = equalize_adapthist(hubble, clip_limit = 0.01, nbins=256)

plt.subplot(1,2,1)
io.imshow(hubble)

plt.subplot(1,2,2)
io.imshow(new_hubble)
```

```python
# Otsu and binary

otsu_hubble = color.rgb2gray(new_hubble)
otsu = threshold_otsu(otsu_hubble)
binary = otsu_hubble > otsu
io.imshow(binary)
```

```python
# Erosion

b = square(12)
new_erosion = erosion(binary, b)
io.imshow(new_erosion)
```

```python
# Remove small objects

removed = remove_small_objects(new_erosion)
io.imshow(removed)
```

```python
# Dilation

b = disk(12)
new_dilation = dilation(removed, b)
io.imshow(new_dilation)
```

```python
# Label

new_label = label(new_dilation)
io.imshow(new_label)
```

```python
# Reshape

new_label = np.reshape(new_label, (872, 1000, 1))
np.shape(new_label)
io.imshow(new_label)
```

```python
# Clip

scaled_data = np.clip(new_label*hubble, 0, 255).astype(np.uint8) # rescale the data to be within [0, 255] for integers
io.imshow(scaled_data, cmap='plasma')
```

```python
# Plot the result

plt.subplot(1,2,1)
io.imshow(hubble)
plt.subplot(1,2,2)
io.imshow(new_label*hubble, cmap='plasma')
```
