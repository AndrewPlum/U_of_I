## Part 1: data modeling with Pandas  

__Problem 1:__ The cafeteria at UI has a data file to store the following information of its customer: (1) phone number, (2) list of items they have purchased, (3) unit price, (4) quantity (5) date of purchase (6) and time of each purchase. See attached image "cafe.png". Generate such a file as a DataFrame in Pandas with $20$ rows (make it as diverse as possible).


```python
# import packages

import pandas as pd
import random
from random import randint
from random import choices
from datetime import date, timedelta, time

# make cafeteria DataFrame

cafeteria_df = pd.DataFrame(columns = ["Phone Number", "List of Items Purchased", "Unit Price", "Quantity", "Date of Purchase", "Time of Each Purchase"])

# declare variables

list_of_purchaseable_items = ["Pizza Slice", "Hamburger", "Chicken Tenders", "Salad", "Taco", "Burrito"]
list_of_phone_numbers = []
start_date = date(2020, 1, 1)
end_date = date(2023, 12, 31)

# fill phone number lists - used for cafeteria and Rec center DataFrames

for i in range(20):
    phone_number = "(" + str(randint(100,999)) + ")-" + str(randint(100,999)) + "-" + str(randint(1000,9999))
    list_of_phone_numbers.append(phone_number)   
    
# fill cafeteria DataFrame

for i in range(20):
    phone_number = list_of_phone_numbers[i]
    list_of_items_purchased = "".join(choices(list_of_purchaseable_items, k=1))
    unit_price = randint(5,10)
    quantity = randint(1,5)
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    date_of_purchase = random_date.strftime('%b %d, %Y')
    random_time = time(random.randint(0, 23), random.randint(0, 59))
    time_of_each_purchase = random_time.strftime('%I:%M %p')
    cafeteria_df.loc[i] = [phone_number, list_of_items_purchased, unit_price, quantity, date_of_purchase, time_of_each_purchase]

# print cafeteria DataFrame
    
cafeteria_df
```


__Problem 2:__ The REC center at UI has a data file to store the following information of its members: (1) phone number, (2) date of visit (3) starting time of visit, (4) ending time of visit. Make such as data file as Pandas' DataFrame with $20$ rows.

```python
# import packages

import pandas as pd
import random
from random import randint
from random import choices
from datetime import date, timedelta, time

# make REC center DataFrame

rec_center_df = pd.DataFrame(columns = ["Phone Number", "Date of Visit", "Starting Time of Visit", "Ending Time of Visit"])

# declare variables

start_date = date(2010, 1, 1)
end_date = date(2023, 12, 31)
number_of_common_phone_numbers = randint(1,19)

# fill REC center DataFrame

for i in range(number_of_common_phone_numbers): # this for loop fills REC DataFrame with random number of common phone numbers
    phone_number = list_of_phone_numbers[i]
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    date_of_visit = random_date.strftime('%b %d, %Y')
    random_time = time(random.randint(0, 23), random.randint(0, 59))
    starting_time_of_visit = random_time.strftime('%I:%M %p')
    random_time = time(random.randint(0, 23), random.randint(0, 59))
    ending_time_of_visit = random_time.strftime('%I:%M %p')
    rec_center_df.loc[i] = [phone_number, date_of_visit, starting_time_of_visit, ending_time_of_visit]
for i in range(20 - number_of_common_phone_numbers):
    phone_number = "(" + str(randint(100,999)) + ")-" + str(randint(100,999)) + "-" + str(randint(1000,9999)) # this for loop fills REC DataFrame with random number of new phone numbers
    random_date = start_date + timedelta(days=random.randint(0, (end_date - start_date).days))
    date_of_visit = random_date.strftime('%b %d, %Y')
    random_time = time(random.randint(0, 23), random.randint(0, 59))
    starting_time_of_visit = random_time.strftime('%I:%M %p')
    random_time = time(random.randint(0, 23), random.randint(0, 59))
    ending_time_of_visit = random_time.strftime('%I:%M %p')
    rec_center_df.loc[i + number_of_common_phone_numbers] = [phone_number, date_of_visit, starting_time_of_visit, ending_time_of_visit]


# print REC center DataFrame
    
rec_center_df
```

__Problem 3:__ The university wants to make a data file that contains the activities of both its Cafeteria and REC center. Write the code to generate such data file as a Pandas' DataFrame.

```python
# merge DataFrames

combined_ui_df = pd.merge(cafeteria_df, rec_center_df, on='Phone Number', how='outer')

# print combined UI DataFrames

combined_ui_df
```

__Problem 4:__ The REC center wants to see the eating habits of its members to make food recommendation. Write the code to combine the data file for such purpose.

```python
# create DataFrame that contains only the rows of REC center members who have eaten at the cafeteria

rec_center_cafeteria_habits_df = pd.merge(cafeteria_df, rec_center_df, on='Phone Number', how='right')

# print cafeteria habits of recenter members

rec_center_cafeteria_habits_df
```

## Part 2: working with external data

__Problem:__ Read the "supermarket_data.csv" file. Read the names of the columns (nicely!). Then answer the following questions (basing on your detailed analysis/codes).



```python
import pandas as pd

df = pd.read_csv("supermarket_data.csv")
df
```

(a) What is the total gross income of each product line?

```python
gross_income_by_product_line = df.groupby("Product line")["gross income"].sum()

print(gross_income_by_product_line)
```

(b) What is the highest-rated product line in the dataset?

```python
ratings_by_product_line = df.groupby("Product line")["Rating"].mean()
highest_rated_product = ratings_by_product_line.idxmax()

print(ratings_by_product_line)
print("\nHighest rated product line: %s" % highest_rated_product)
```

(c) Which gender spend more money on Electronic accessories?

```python
electronic_accessories_df = df[df["Product line"] == "Electronic accessories"]
electronic_accessories_sold_by_gender_df = electronic_accessories_df.groupby("Gender")["Total"].sum()
which_gender_buys_more = electronic_accessories_sold_by_gender_df.idxmax()

print(electronic_accessories_sold_by_gender_df)
print("\nWhich gender buys more electronic accessories: %s" %which_gender_buys_more)
```

(e) What month in the year that Health and beauty is most consumed?

```python
# Create data frame with months column

months = []
for date in df["Date"]:
    data = str.split(date,'/')
    months.append(data[0])
df['Months'] = months

# Find month health and beauty product line was the most consumed

health_and_beauty_df = df[df["Product line"] == "Health and beauty"]
health_and_beauty_by_month_df = health_and_beauty_df.groupby("Months")["Total"].sum()
month_health_and_beauty_most_consumed = health_and_beauty_by_month_df.idxmax()

# Print tables and answer

print(health_and_beauty_by_month_df)
print("\nWhich month in the year is health and beauty most consumed: %s" %month_health_and_beauty_most_consumed)
```

(g) Study the relationships between the following quantities:'Unit price', 'Quantity', 'Tax 5%','cogs','gross margin percentage','gross income' (state at least three relationships and plot them). 

```python
import matplotlib.pyplot as plt

#(I): Unit Price and Gross Margin Percentage

plt.scatter(df["Unit price"], df["gross margin percentage"])
plt.xlabel("Unit Price")
plt.ylabel("Gross Margin Percentage")
plt.show()
print("Relationship: There is no correlation between the unit price and the gross margin percentage.")

#(II): Tax 5% and Costs of Goods Sold

plt.scatter(df["cogs"], df["Tax 5%"])
plt.xlabel("Costs of Goods Sold")
plt.ylabel("Tax 5%")
plt.show()
print("Relationship: There is a positive correlation between the costs of goods sold and the tax being 5%.")
print("Formula: Tax 5% = Costs of Goods Sold * 0.05")

#(III): Tax 5% and Gross Margin Percentage

plt.scatter(df["Tax 5%"], df["gross margin percentage"])
plt.xlabel("Tax 5%")
plt.ylabel("Gross Margin Percentage")
plt.show()
print("Relationship: There is no correlation between the tax being 5% and the gross margin percentage.")

#(IV): Tax 5% and Gross Income

plt.scatter(df["gross income"], df["Tax 5%"])
plt.xlabel("Gross Income")
plt.ylabel("Tax 5%")
plt.show()
print("Relationship: There is a positive correlation between the gross income and the tax being 5%.")
print("Formula: Tax 5% = Gross Income")

#(V): Costs of Goods Sold and Gross Margin Percentage

plt.scatter(df["cogs"], df["gross margin percentage"])
plt.xlabel("Costs of Goods Sold")
plt.ylabel("Gross Margin Percentage")
plt.show()
print("Relationship: There is no correlation between the costs of goods sold and the gross margin percentage.")

#(VI): Costs of Goods Sold and Gross Income

plt.scatter(df["cogs"], df["gross income"])
plt.xlabel("Costs of Goods Sold")
plt.ylabel("Gross Income")
plt.show()
print("Relationship: There is a positive correlation between the costs of goods sold and the gross income.")
print("Formula: Gross Income = Costs of Goods Sold * 0.05")

#(VI): Gross Income and Gross Margin Percentage

plt.scatter(df["gross income"], df["gross margin percentage"])
plt.xlabel("Gross Income")
plt.ylabel("Gross Margin Percentage")
plt.show()
print("Relationship: There is no correlation between the gross income and the gross margin percentage.")

print()
print("Conclusions: ")
print("1) Gross margin percentage is constant and is independent of other variables.")
print("2) Formula: Tax 5% = Costs of Goods Sold * 0.05")
print("3) Formula: Tax 5% = Gross Income")
print("4) Formula: Gross Income = Costs of Goods Sold * 0.05")
```
