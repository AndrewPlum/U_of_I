## 2D vs 3D ploting

We can plot a function $z = f(x,y)$ in three dimensional space.

```python
import numpy as np
import matplotlib.pyplot as plt
```

__Problem:__ Plot the function $f(x,y) = \sin(\sqrt{x^2+y^2})$ in the square $[-6,6] \times [-6,6]$. 

```python
# Define the function we want to plot 

# Generate the domain of your function
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)
X, Y = np.meshgrid(x,y)

# Generate your data
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))

Z = f(X, Y)
Z.shape
```

```python
# 2D plot with colorbar 

plt.figure(figsize = (3,3))
plt.imshow(Z,cmap='plasma') #cmap stands for color map - can do viridis, reds, plasma, etc
plt.colorbar() # the color bar will be between 0 and 1 because of the sin function
plt.show()
```

```python
# 3D plot using contour 

%matplotlib notebook

fig = plt.figure(figsize=(3,3))
ax = plt.axes(projection='3d')

#note: play around with color map
ax.contour3D(X, Y, Z, 100, cmap='plasma') # contour stays at the same z value
```

```python
# 3D plot using surface 
plt.figure(figsize = (3,3))
ax2 = plt.axes(projection='3d')

ax2.plot_surface(X,Y,Z, cmap = "plasma")
```

## Time depedent plotting

__Problem:__ Given the function $f(x,t) = \sin(x+t)$ for $x \in [-\pi,\pi]$ and $t \in [0,2 \pi]$. Plot the function $f$ in time.

```python
from numpy import pi
x = np.linspace(-pi,pi,100)
t = np.linspace(0,2*pi,50)

# Generate the meshgrid and function
X,T = np.meshgrid(x,t)

plt.figure(figsize=(3,3))
Z = np.sin(X+T) # can also define using function definition
plt.imshow(Z)
```

```python
# Static plots

plt.figure(figsize=(3,3))
plt.plot(Z[0])
plt.show()

plt.figure(figsize=(3,3))
plt.plot(Z[25])
plt.show()
```

```python
#Dynamic plot - animation
from matplotlib import animation

# First set up the figure, the axis, and the plot element we want to  animate
fig = plt.figure(figsize=(3,3))
ax = plt.axes(xlim=(-np.pi, np.pi), ylim=(-2, 2))

#plot the intial function
line, = ax.plot([], [], lw=2)

# animation function.  This is called sequentially

def animate(i):
    x = np.linspace(-np.pi, np.pi, 100)
    y = Z[i]
    line.set_data(x,y)
    return line

anim = animation.FuncAnimation(fig, animate,
                           frames=200, interval=30, blit=True)
```
