__Problem:__

(a) Plot the helix $x=\sin(t), y = \cos(t), z=t$, $t =0,10 \times pi$.

```python
# 3D animation
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D

# Create the data for the spiral
t = np.linspace(0, 10*np.pi, 1000)
x = np.sin(t)
y = np.cos(t)
z = t

# Create a figure and 3D axes
fig = plt.figure() # done so the pictures don't mush together
ax = plt.axes(projection='3d')

# Create an initial plot of the spiral
ax.plot(x,y,z,lw=2)
```

(b) Rotate it around the $z$-axis.

```python
# 3D animation
import numpy as np
from matplotlib import animation
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
# Create the data for the spiral
t = np.linspace(0, 10*np.pi, 1000)
x = np.sin(t)
y = np.cos(t)
z = t

# Create a figure and 3D axes
fig = plt.figure(figsize=(3,3)) # done so the pictures don't mush together
ax = plt.axes(projection='3d')

# DIFFERENT PART FROM ABOVE FOR 3D PICTURE

# Create an initial plot of the spiral
line, = ax.plot(x,y,z,lw=2) # "line,"" is tuple unloading

# Define the animation function
def animate(i):
    ax.view_init(elev=10., azim=i)
    return line,

# Create the animation object
ani = animation.FuncAnimation(fig, animate, frames=360, interval=50)

# Show the animation
plt.show()
```

__Problem:__ 

(a) Plot the function $f(x,y) = \sin(xy)$ in $[-2 \pi,2 \pi] \times [-2 \pi,2 \pi]$.

```python
x=y = np.linspace(-2* np.pi,2*np.pi,100)
X,Y = np.meshgrid(x,y)
f = np.sin(X * Y)
plt.figure(figsize=(3,3)) # done so the pictures don't mush together
plt.imshow(f,cmap = 'plasma')
```

(b) Plot the animation for $f(x,y)=sin(xy +t)$ in $[-2 \pi,2 \pi] \times [-2 \pi,2 \pi]$ for $t \in [0,5]$.

```python
from matplotlib import animation
x=y = np.linspace(-2* np.pi,2*np.pi,100)
t = np.linspace(0,5,100)
X,Y,T = np.meshgrid(x,y,t) # X, Y, T are rank 3 tensor
f = np.sin(X*Y+T)
fig = plt.figure(figsize=(3,3)) # done so the pictures don't mush together
ax = plt.imshow(f[:,:,0],cmap = 'plasma')

def animate(i):
    ax.set_data(f[:,:,i]) # i is the time dimension in time
    return ax,

# Create the animation object
ani = animation.FuncAnimation(fig, animate, frames=360, interval=50)

# Show the animation
plt.show()
```

__Practice:__ Let $\vec{v}=(\cos \pi/3, \sin \pi/3)$. 

(a) Plot the function $f(x,y) = \sin((x,y) \cdot \vec{v})$ for $(x,y) \in [-2\pi, 2\pi] \times [-2\pi, 2\pi]$.

```python
from matplotlib import animation

angle = np.pi/3
x=y = np.linspace(-2* np.pi,2*np.pi,100)
X,Y = np.meshgrid(x,y) # X and Y are matrices
v = np.array([np.cos(angle), np.sin(angle)])
dot = v[0]*X + v[1]*Y
f = np.sin(dot)
fig = plt.figure(figsize=(3,3))
plt.imshow(f)
```

(b) Plot the traveling wave function $f(x,y) = \sin((x,y) \cdot \vec{v}-t)$ for $(x,y) \in [-2\pi, 2\pi] \times [-2\pi, 2\pi]$ and $t \in [0,10]$.

```python
angle = np.pi
x=y = np.linspace(-2* np.pi,2*np.pi,100)
t = np.linspace(0,10,100)
X,Y,T = np.meshgrid(x,y,t) # X, Y, T are rank 3 tensor???
v = np.array([np.cos(angle), np.sin(angle)])
dot = v[0]*X + v[1]*Y
f = np.sin(dot-T)
fig = plt.figure(figsize=(3,3))
ax = plt.imshow(f[:,:,0])

# Define the animation function
def animate(i):
    ax.set_data(f[:,:,i])
    return ax,

# Create the animation object
ani = animation.FuncAnimation(fig, animate, frames=360, interval=50)

# Show the animation
plt.show()
```
