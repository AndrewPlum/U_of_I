```python
import pandas as pd

# Creating a sample dataframe
df = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar'],
                   'B': ['one', 'one', 'two', 'one'],
                   'C': [1, 2, 3, 4],
                   'D': [10, 20, 30, 40]})

# Grouping the dataframe by columns 'A' and 'B'
grouped = df.groupby(['A', 'B'])

# Calculating the sum of the values in columns 'C' and 'D' for each group
result = grouped.sum()

print("Grouped Result:")
print(result)
```

In the example above, the __groupby__ method is used to group the data in the DataFrame df based on the values in columns A and B. The resulting DataFrame grouped is then used to calculate the __sum__ of the values in columns C and D for each group.

The resulting DataFrame result contains the sum of the values in columns C and D for each unique combination of values in columns A and B. The resulting DataFrame is indexed by the unique combinations of values in columns A and B.
