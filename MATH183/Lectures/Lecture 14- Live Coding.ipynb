# Random module
Random module can be used to generate data randomly, obeying some specific rules.

Exmple:

```Python
    from random import random
    random()
```
This randomly chooses a number in $[0,1)$. 

````python
# gerate a set list of 1000 random numbers in [0,1) 
# Never generates 1
# Uniform distribution -every number between 0 inclusive and 1 have an equal chance of generating

from random import random

random_list = []
for i in range(10000):
    random_list.append(random())

#random()
````

````python
# plot the histogram
# histogram: plot frequency vs values 

import matplotlib.pyplot as plt

plt.hist(random_list, bins = 100) # bins makes 100 intervals; the default is 10
plt.show()
````

````python
# random generator using numpy

from numpy.random import random

random_arr = random(1000)
plt.hist(random_arr, bins = 100)
````

## random.normal

Use the random.normal() method to get a Normal Data Distribution. It has three parameters:

- loc - (Mean) where the peak of the bell exists.

- scale - (Standard Deviation) how flat the graph distribution should be.

- size - The shape of the returned array.



````python
# generate a normal distribution 

from numpy.random import normal

random_array = normal(0,10,10000)
plt.hist(random_array, bins = 100)
````

# pivot table 
Arrange data in a table (using index and columns). Values of one or more column(s) are used as index, values of one or more column(s) are used as name of columns.

````python
# Pivot table - two dimensional version of groupby basically

import pandas as pd

df = pd.read_csv('supermarket_data.csv')
#df.head()
````

__Problem:__ Generate a pivot table using 'Branch' as index and 'Gender' as columns to compute the gross income.

````python
table = df.pivot_table(index = "Branch", columns = "Gender", values = "gross income", aggfunc = "sum")
table
````
