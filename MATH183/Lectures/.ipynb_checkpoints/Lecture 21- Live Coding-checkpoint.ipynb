## Gray scale image

- Values between $0$ (black) and $1$ (white) if the values are __floats__.

- Values between $0$ (black) and $255$ (white) if the value are __integers__.

## Color images

A color image normally has three channels: Red, Green, Bule.

## Conversion between grayscale and color image

- Converting grayscale to color: skimage.color.gray2rgb(image, *, channel_axis=-1)

- Converting color to grayscale: skimage.color.rgb2gray(rgb, *, channel_axis=-1)

__NOTE:__ Formula of conversion from rgb to gray: 
$$Y = 0.2125 R + 0.7154 G + 0.0721 B.$$

```python
# gray to color
from skimage import color,data, io
camera = data.camera()
io.imshow(camera)
camera.shape
camera.dtype
```

```python
color_camera = color.gray2rgb(camera)
#print(color_camera.shape)
io.imshow(color_camera)
```

```python
blue_camera = color_camera.copy()
blue_camera[:,:,:2] =0
io.imshow(blue_camera)
```

```python
# color to gray
immu = data.immunohistochemistry()
io.imshow(immu)
```

```python
# color to gray
gray_immu = color.rgb2gray(immu)
gray_immu.shape
#io.imshow(gray_immu)
```

```python
# Binarization 
# Makes everything close to black completely black and 
# stuff closer to white becomes white
# Basically a way to decompress an image

#io.imshow(camera)
threshold = 120 # 120 works as a good threshold - look at histogram for valley
binary = camera > threshold
binary.shape # binary should be a matrix
io.imshow(binary)
```

```python
# Histogram 
import matplotlib.pyplot as plt

plt.hist(camera.ravel(), bins = 100)# Histogram
```

```python
# Thresholding & plotting 

thresh1 = 0.74 # optimal value because it corresponds to a value on the histogram above
binary2 = gray_immu > thresh1
plt.subplot(1,2,1)
io.imshow(gray_immu)
plt.subplot(1,2,2)
io.imshow(binary2)
```

```python
#Otsu's threshoding
# Picks an "optimal" threshold value for binarization
from skimage.filters import threshold_otsu

thresh = threshold_otsu(camera)
binary3 = camera > thresh
io.imshow(binary3)
```

```python
# nicer plot

fig, axes = plt.subplots(ncols=3, figsize=(8, 2.5))
ax = axes.ravel()
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

ax[0].imshow(camera, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].hist(camera.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(thresh, color='r')

ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2].set_title('Thresholded')
ax[2].axis('off')

plt.show()
```

## More Datasets


Skimage also provides image [data sets](https://scikit-image.org/docs/stable/api/skimage.data.html) and sample images to experiment with, making it easy for users to get started with image processing.



```python
# coins
from skimage import filters, feature

coins = data.coins()
io.imshow(coins)

# edge detection

edge = filters.sobel(coins)
plt.subplot(1,3,1)
io.imshow(coins)
plt.subplot(1,3,2)
io.imshow(edge)

edge_canny = feature.canny(coins, sigma = 3, low_threshold = 10, high_threshold = 80)
plt.subplot(1,3,3)
io.imshow(edge_canny)
```
