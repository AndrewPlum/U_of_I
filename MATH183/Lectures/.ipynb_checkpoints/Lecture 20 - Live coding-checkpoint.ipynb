# Scikit-image

- __Scikit-image__ (skimage) is an open-source image processing library that provides a collection of algorithms for image processing and computer vision tasks. 

- It is built on top of the NumPy and SciPy libraries, and it provides simple and efficient tools for image manipulation, enhancement, restoration, segmentation, feature extraction, and more.

## Gray scale image

- Values between $0$ (black) and $1$ (white) if the values are __floats__.

- Values between $0$ (black) and $255$ (white) if the value are __integers__.

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage import io
x=y = np.linspace(-2,2,1000)
X,Y = np.meshgrid(x,y)
f = np.zeros(X.shape)
f[X**2 + Y**2 < 1] = 0.95
io.imshow(f)
```

```python
from skimage import data
camera = data.camera()
height, width = camera.shape[0], camera.shape[1] # o is vertical and 1 is horizontal dimensions
io.imshow(camera)
idx = camera.argmax()
x_max, y_max = idx % width, idx/width
print(x_max, y_max)
```

## Color images

A color image normally has three channels: Red, Green, Bule.

```python
# red image
red_image = np.zeros([X.shape[0],X.shape[1],3])
red_image[:,:,0] = f
io.imshow(red_image)
```

```python
# green image
green_image = np.zeros([X.shape[0],X.shape[1],3])
green_image[:,:,1] = f
io.imshow(green_image)
```

```python
#Blue image
blue_image = np.zeros([X.shape[0],X.shape[1],3])
blue_image[:,:,2] = f
io.imshow(blue_image)
```

```python
#Combining the colors 

io.imshow(red_image/1.5 + blue_image) # you can weight it
```

```python
# saving an image 

io.imsave("blue.png", blue_image)
```

```python
# loading an image

load_blue = io.imread("blue.png")
load_blue.dtype
```

## Conversion between grayscale and color image

- Converting grayscale to color: skimage.color.gray2rgb(image, *, channel_axis=-1)

- Converting color to grayscale: skimage.color.rgb2gray(rgb, *, channel_axis=-1)

__NOTE:__ Formula of conversion from rgb to gray: 
$$Y = 0.2125 R + 0.7154 G + 0.0721 B.$$

```python
# gray to color
from skimage import color

color_camera = color.gray2rgb(camera)
color_camera.shape
io.imshow(color_camera)
```

```python
# Play around with color channels
from skimage import color

blue_camera = color_camera.copy()
blue_camera[:,:,2] = 0
io.imshow(blue_camera)
```
