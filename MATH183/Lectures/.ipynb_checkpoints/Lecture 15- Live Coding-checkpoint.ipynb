## Boxplot

A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It shows the median, quartiles, and outliers of the data.

In Python, you can create boxplots using the matplotlib or seaborn

```python
# boxplot with seaborn
# boxplot properties
# the box contains 50%
# the whiskers are the intervals outsitde of the box of the boxplot whose maximum length are 1.5 times the length of the box but stops if values not in that full range
# points outside the whiskers are considered outliers - used for anomally detection

sns.boxplot(x = "Gender", y = "gross income", data = df)
plt.show()
```

The resulting plot will show a box with the median line, the upper and lower quartiles (the 25th and 75th percentiles), and whiskers that extend to the most extreme data points that are within 1.5 times the interquartile range (IQR) of the upper and lower quartiles. Any data points outside of this range are considered outliers and are plotted as individual points.

### How to calculate lower/upper adjacent value?

To calculate the lower adjacent value (also known as the lower whisker) in a box plot, you need to first find the first quartile (Q1) and the interquartile range (IQR) of your data.

- Find the first quartile (Q1), which is the median of the lower half of your data. To do this, first sort your data from smallest to largest, then find the median of the lower half of your data. If your data set has an odd number of values, exclude the median from both halves.

- Find the interquartile range (IQR), which is the difference between the third quartile (Q3) and the first quartile (Q1). To do this, find Q3 using the same method as Q1, but for the upper half of your data.

- Calculate the lower adjacent value as follows:

 Lower adjacent value = Q1 - 1.5 x IQR

- Any data points that fall below the lower adjacent value are considered outliers and are plotted as individual points outside the whiskers in the box plot.

## Violin plot

A violin plot is another type of data visualization that shows the distribution of a dataset. It is similar to a boxplot, but instead of showing only summary statistics like the median and quartiles, it shows the kernel density estimation of the underlying distribution. This makes it useful for visualizing the shape and density of the data.

In Python, you can create violin plots using the violinplot function from the seaborn library, which is a popular data visualization library built on top of matplotlib

```python
# violin plot
# like box plot it shows the concentration of data

sns.violinplot(x='Gender', y='gross income', data = df)
plt.show()
```

The resulting plot will show a violin shape with the kernel density estimation of the data. The wider parts of the plot indicate regions of higher density, and the narrower parts indicate regions of lower density. The white dot in the middle of the plot represents the median of the data, and the thin black lines on either side represent the interquartile range (IQR).
