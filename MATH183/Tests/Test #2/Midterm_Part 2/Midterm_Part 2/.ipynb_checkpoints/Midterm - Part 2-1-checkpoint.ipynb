__Probem 2: (40 points)__ 

(a) Read the data from the file "Simplified orders.csv" as a DataFrames named df_orders.

```python
import pandas as pd

# Read in Simplified orders.csv file as a DataFrame
df_orders = pd.read_csv('Simplified orders.csv')
print('Simplified orders.csv:')
display(df_orders.head(10))
```

(b) Make a DataFrame that only contains the orders from the restaurant "R10002" or "R10003".

```python
restaurant_orders = df_orders[(df_orders['RESTAURANT_ID'] == 'R10002') | (df_orders['RESTAURANT_ID'] == 'R10003')]
restaurant_orders
```

(c) Make a DataFrame that only contains the orders which have more than one cookies.

```python
cookie_drink_orders = df_orders[(df_orders['COOKIES'] > 0)]
cookie_drink_orders
```

(d) Create a column named "MONTH" that contain the month of each order.

```python
df_orders["MONTH"] = pd.to_datetime(df_orders["DATETIME"]).dt.month
df_orders
```

(e) Find the restaurant that sold most cookies.

```python
df_orders.groupby("RESTAURANT_ID")["COOKIES"].sum().idxmax()
```

(f) Make a table to summarize the total number of drinks sold in terms of restaurant ID and month.

```python
table = df_orders.pivot_table(index = "RESTAURANT_ID", columns = "MONTH", values = "DRINKS", aggfunc = "sum")
table
```

(a) Read the data from "weather.csv" as a DateFrame named df_weather. Plot a line chart for the temperature, using the datatime as x values.

```python
import pandas as pd 
import seaborn as sns

df_weather = pd.read_csv("weather.csv")
df_weather
#df_weather["DATE"] = pd.to_datetime(df_weather["DATETIME"]).dt.date
#df_weather
temperature = df_weather.groupby("DATETIME")["TEMPERATURE"].mean()
#df_weather.plot(kind = "line", x = df_weather["DATETIME"], y = df_weather["TEMPERATURE"])
#temperature
sns.lineplot(data = temperature)
```

(b) Compute the average temperature for each month. Then plot it using a bar plot.

```python
import matplotlib.pyplot as plt

df_weather["MONTH"] = pd.to_datetime(df_weather["DATETIME"]).dt.month
weather_month = df_weather.groupby("MONTH")["TEMPERATURE"].mean()
#display(df_weather)
#weather_month
plt.bar(weather_month.index, weather_month)
```

(c)Using multiple boxplot to summarize the temperature for each month.

```python
temp_date = weather_month.groupby("MONTH" == 1)["TEMPERATURE"]
plt.boxplot(temp_date)
plt.show()

#temp_date = df_weather.groupby("MONTH")["TEMPERATURE"].mean()
#plt.boxplot(temp_date)
#plt.show()
```
