__Probem 1:__ Read the data from the file "Simplified order.csv" as a DataFrames named df_orders.



````python
import pandas as pd
path = 'raw data MBAs/'
df_orders = pd.read_csv(path+'Simplified orders.csv')
````

(a) Plot the histogram of the columns DRINKS using Mathplotlib, Seaborn, and Pandas, respectively. NOTE: you may want to take log scale to make it look nice.

#### Histogram
 - Matplotlib: 
    ```Python     
     plt.hist(data)
    ```
    Here, 
      - plt stands for matplotlib.pylot
      - data: an array that contains values
 - Seaborn: 
     ```Python
     sns.barplot(data)
     ```
   Here,
      - sns stands for seaborn
      - data is an array that contains values
      
 - Pandas: 
     ```Python
    df.plot(kind = 'hist', column= column_name)
    ```
    Here,
      - df stands for a DataFrame
      - column_name is the name of column that has the data

````python
import matplotlib.pyplot as plt
plt.hist(df_orders['DRINKS'],log=True, bins = 23)
plt.show()
````

````python
import seaborn as sns
sns.histplot(df_orders['DRINKS'],log=True)
````

````python
import numpy as np
df_orders.plot(kind='hist',column = 'DRINKS', bins= 23, logy=True)
````

(b) Plot the number of drinks sold each date as a line chart.

#### Line plot/chart:

 - Matplotlib: 
    ```Python     
     plt.plot(xdata,ydata)
    ```
    Here, 
      - plt stands for matplotlib.pylot
      - xdata: an array that contains x values
      - ydata: an array that contains y values
      - NOTE: you can input ONE data array instead, then it is the ydata array.
      
      
 - Seaborn: 
     ```Python
     sns.lineplot(x=xdata,y=ydata)
     ```
   Here,
      - sns stands for seaborn
      - xdata is an array that contains x values
      - ydata is an array that contains y values
      - NOTE: if you only has one array (ydata) and plot its values as a line chart, then the syntax is
    ```Python
     sns.lineplot(ydata)
     ```

 - Pandas: 
     ```Python
    df.plot(kind = 'line', x= columnx, y = columny)
    ```
    Here,
      - df stands for a DataFrame
      - columnx, columy are the names of the columns of x and y data, respectively.
      - if df is a Series (one column of data), then the syntax is:
     ```Python
    df.plot(kind = 'line')
    ```
      

````python
df_orders['DATE'] = pd.to_datetime(df_orders['DATETIME']).dt.date
df_sum_drinks = df_orders.groupby('DATE')['DRINKS'].sum()
````

````python
type(df_sum_drinks)
#plt.plot(df_sum_drinks.index,df_sum_drinks.values)
````

````python
sns.lineplot(df_sum_drinks.values)
````

````python
df_sum_drinks.plot(kind='line')
````

(c) Barplot for sum of drinks per month.

#### Bar plot

 - Matplotlib: 
    ```Python     
     plt.bar(xdata,ydata)
    ```
    Here, 
      - plt stands for matplotlib.pylot
      - xdata: an array that contains x values
      - ydata: an array that contains y values
 - Seaborn: 
     ```Python
     sns.barplot(x=xdata,y=ydata)
     ```
   Here,
      - sns stands for seaborn
      - xdata is an array that contains x values
      - ydata is an array that contains y values

 - Pandas: 
     ```Python
    df.plot(kind = 'bar', x= columnx, y = columny)
    ```
    Here,
      - df stands for a DataFrame
      - columnx, columy are the names of the columns of x and y data, respectively.
      - NOTE: if your DataFrame only has one column of data, you don't have to specify the columns. Pandas will automatically use the index as x values and the data as y values. 

````python
df_orders['MONTH'] = pd.to_datetime(df_orders['DATETIME']).dt.month
````

````python
df_sum_month = df_orders.groupby('MONTH')['DRINKS'].sum()
plt.bar(df_sum_month.index,df_sum_month)
````

````python
sns.barplot(data=df_sum_month, x=df_sum_month.index,y=df_sum_month)
````

````python
df_sum_month.plot(kind='bar')
````

(f) Compute the total cookies for each month. Do the scatter plot of this quantity vs the total drinks each month.


#### Scatter plot
 - Matplotlib: 
    ```Python     
     plt.scatter(xdata,ydata)
    ```
    Here, 
      - plt stands for matplotlib.pylot
      - xdata: an array that contains x values
      - ydata: an array that contains y values
 - Seaborn: 
     ```Python
     sns.scatterplot(x=xdata,y=ydata)
     ```
   Here,
      - sns stands for seaborn
      - xdata is an array that contains x values
      - ydata is an array that contains y values

 - Pandas: 
     ```Python
    df.plot(kind = 'scatter', x= columnx, y = columny)
    ```
    Here,
      - df stands for a DataFrame
      - columnx, columy are the names of the columns of x and y data, respectively.

````python
df_cookies_sum = df_orders.groupby('MONTH')['COOKIES'].sum()
plt.scatter(df_cookies_sum,df_sum_month)
plt.xlabel('Cookies')
plt.ylabel('Drinks')
plt.show()
````

````python
sns.scatterplot(x=df_cookies_sum,y = df_sum_month)
plt.xlabel('Cookies')
plt.ylabel('Drinks')
plt.show()
````

````python
df_drink_cookie = df_orders.groupby('MONTH')['COOKIES','DRINKS'].sum()
df_drink_cookie.plot(kind='scatter',x='COOKIES', y = 'DRINKS',c='r')
````

(g) (Multiple box/violin plot) Make a new DataFrame that contains all orders which have more than 5 drinks. Use the boxplot to describe the number of drinks each month.


#### Boxplot

__(b) Multiple boxes:__

 - Matplotlib: 
    ```Python     
     plt.box([data])
    ```
    Here, 
      - plt stands for matplotlib.pylot
      - [data] is a list of data array
      
 - Seaborn: 
     ```Python
     sns.boxplot(x=columnx, y =columny, data = df)
     ```
   Here,
    - df is a DataFrame
    - columnx: name is the column that contains the x-values
    - columny: name is the column that contains the y-values
      
 - Pandas: 
     ```Python
    df.plot(kind = 'box', column= [column_names])
    ```
    Here,
      - df stands for a DataFrame
      - [column_name] is the list of column names.

````python
df_orders_with_drink = df_orders[df_orders['DRINKS']>5]
````

````python
sns.boxplot(x='MONTH', y ='DRINKS', data = df_orders_with_drink)
plt.show()
````
