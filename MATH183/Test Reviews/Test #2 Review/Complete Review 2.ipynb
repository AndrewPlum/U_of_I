## Pandas Groupby

````python
import pandas as pd

# create sample data
data = {'date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02', '2022-01-03'],
        'product': ['A', 'B', 'A', 'B', 'A'],
        'quantity': [10, 20, 5, 15, 10],
        'price': [1.50, 2.00, 1.75, 2.25, 1.50]}
df = pd.DataFrame(data)

# group by product and sum the revenue
revenue_by_product = df.groupby('product')['quantity', 'price'].sum()
revenue_by_product['revenue'] = revenue_by_product['quantity'] * revenue_by_product['price']

print(revenue_by_product)
````

## Pandas Pivot Table

````python
import pandas as pd

# create sample data
data = {'date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02', '2022-01-03'],
        'product': ['A', 'B', 'A', 'B', 'A'],
        'quantity': [10, 20, 5, 15, 10],
        'price': [1.50, 2.00, 1.75, 2.25, 1.50]}
df = pd.DataFrame(data)

# create pivot table
revenue_by_date_and_product = pd.pivot_table(df, values='quantity', index='date', columns='product', aggfunc='sum')
revenue_by_date_and_product = revenue_by_date_and_product.mul(df.groupby(['date', 'product'])['price'].first(), axis=1)
revenue_by_date_and_product = revenue_by_date_and_product.sum(axis=1).to_frame(name='revenue')

print(revenue_by_date_and_product)
````
