```python
# a

import pandas as pd

# Read in Simplified orders.csv file as a DataFrame
simplified_orders_df = pd.read_csv('Simplified orders.csv')
print('Simplified orders.csv:')
display(simplified_orders_df.head(10))

# Read in Summarized orders.csv file as a DataFrame
summarized_orders_df = pd.read_csv('Summarized orders.csv')
print('Summarized orders.csv:')
display(summarized_orders_df.head(10))

# Read in weather.csv file as a DataFrame
weather_df = pd.read_csv('weather.csv')
print('weather.csv:')
display(weather_df.head(10))

# Read in Items.csv file as a DataFrame
items_df = pd.read_csv('Items.csv')
print('Items.csv:')
display(items_df.head(10))

# Read in Restaurants.csv file as a DataFrame
restaurants_df = pd.read_csv('Restaurants.csv')
print('Restaurants.csv:')
display(restaurants_df.head(10))
```

```python
# b

# Print out basic information of Simplified orders.csv
print("Simplified orders.csv:")
print(simplified_orders_df.info())

# Print out basic information of Summarized orders.csv
print("Summarized orders.csv:")
print(summarized_orders_df.info())

# Print out basic information of weather.csv
print("weather.csv:")
print(weather_df.info())

# Print out basic information of Items.csv
print("Items.csv:")
print(items_df.info())

# Print out basic information of Restaurants.csv
print("Restaurants.csv:")
print(restaurants_df.info())
```

```python
# c

# Print out description of Simplified orders.csv
print("Simplified orders.csv:")
print(simplified_orders_df.describe())

# Print out description of Summarized orders.csv
print("Summarized orders.csv:")
print(summarized_orders_df.describe())

# Print out description of weather.csv
print("weather.csv:")
print(weather_df.describe())

# Print out description of Items.csv
print("Items.csv:")
print(items_df.describe())

# Print out description of Restaurants.csv
print("Restaurants.csv:")
print(restaurants_df.describe())
```

```python
# d 

# Drop rows with NaN values in Simplified orders.csv
simplified_orders_dropped_df = simplified_orders_df.dropna()
print("Simplified orders.csv:")
print(simplified_orders_dropped_df.info())
print("Number of rows dropped:", len(simplified_orders_df) - len(simplified_orders_dropped_df))

# Drop rows with NaN values in Summarized orders.csv
summarized_orders_dropped_df = summarized_orders_df.dropna()
print("Summarized orders.csv:")
print(summarized_orders_dropped_df.info())
print("Number of rows dropped:", len(summarized_orders_df) - len(summarized_orders_dropped_df))

# Drop rows with NaN values in weather.csv
weather_dropped_df = weather_df.dropna()
print("weather.csv:")
print(weather_dropped_df.info())
print("Number of rows dropped:", len(weather_df) - len(weather_dropped_df))

# Drop rows with NaN values in Items.csv
items_dropped_df = items_df.dropna()
print("Items.csv:")
print(items_dropped_df.info())
print("Number of rows dropped:", len(items_df) - len(items_dropped_df))

# Drop rows with NaN values in Restaurants.csv
restaurants_dropped_df = restaurants_df.dropna()
print("Restaurants.csv:")
print(restaurants_dropped_df.info())
print("Number of rows dropped:", len(restaurants_df) - len(restaurants_dropped_df))
```

```python
# e

# Drop columns with NaN values in Simplified orders.csv
simplified_orders_dropped_cols_df = simplified_orders_df.dropna(axis=1)
print("Simplified orders.csv:")
print(simplified_orders_dropped_cols_df.info())
print("Number of columns dropped:", len(simplified_orders_df.columns) - len(simplified_orders_dropped_cols_df.columns))

# Drop columns with NaN values in Summarized orders.csv
summarized_orders_dropped_cols_df = summarized_orders_df.dropna(axis=1)
print("Summarized orders.csv:")
print(summarized_orders_dropped_cols_df.info())
print("Number of columns dropped:", len(summarized_orders_df.columns) - len(summarized_orders_dropped_cols_df.columns))

# Drop columns with NaN values in weather.csv
weather_dropped_cols_df = weather_df.dropna(axis=1)
print("weather.csv:")
print(weather_dropped_cols_df.info())
print("Number of columns dropped:", len(weather_df.columns) - len(weather_dropped_cols_df.columns))

# Drop columns with NaN values in Items.csv
items_dropped_cols_df = items_df.dropna(axis=1)
print("Items.csv:")
print(items_dropped_cols_df.info())
print("Number of columns dropped:", len(items_df.columns) - len(items_dropped_cols_df.columns))

# Drop columns with NaN values in Restaurants.csv
restaurants_dropped_cols_df = restaurants_df.dropna(axis=1)
print("Restaurants.csv:")
print(restaurants_dropped_cols_df.info())
print("Number of columns dropped:", len(restaurants_df.columns) - len(restaurants_dropped_cols_df.columns))
```

__Probem 3:__ Read the data from the file "Simplified order.csv" as a DataFrames named orders.

(a) Make a DataFrame that only contains the orders from the restaurant "R10002".

(b) Make a DataFrame that only contains the orders which have cookie(s) and drink(s).

(c) How many different restaurant are there in the file?

(d) How many types of order are there in the file?

(e) Create a column named "DATE" that contain the date of each order. Hint: extracting Date from the "DATETIME" column using pd.to_datetime function.

(f) Create a column named "DAY" that contain the day (during the weel) of each order. Hint: extracting Date from the "DATETIME" column using pd.to_datetime function.

(g) Determine how many year(s) does the dataset span.


```python
import pandas as pd

# Read in the data as a DataFrame
df_orders = pd.read_csv("Simplified orders.csv")
```

```python
# a 

# Filter the DataFrame to only include orders from R10002
restaurant_orders = df_orders[df_orders['RESTAURANT_ID'] == 'R10002']
restaurant_orders
```

```python
# b

# Filter the DataFrame to only include orders with cookies and drinks
cookie_drink_orders = df_orders[(df_orders['COOKIES'] > 0) & (df_orders['DRINKS'] > 0)]
cookie_drink_orders
```

```python
# c 

# Determine the number of different restaurants in the file
num_restaurants = df_orders['RESTAURANT_ID'].nunique()
print(num_restaurants)
```

```python
# d

# Determine the number of different types of order in the file
num_order_types = df_orders['TYPE'].nunique()
print(num_order_types)
```

```python
# e 

# Convert the "DATETIME" column to a datetime format
df_orders['DATETIME'] = pd.to_datetime(df_orders['DATETIME'])

# Create a new "DATE" column
df_orders['DATE'] = df_orders['DATETIME'].dt.date

df_orders
```

```python
# f

# Create a new "DAY" column
df_orders['DAY'] = df_orders['DATETIME'].dt.day_name()
df_orders
```

```python
# g

# Determine how many years the dataset spans
min_date = df_orders['DATE'].min()
max_date = df_orders['DATE'].max()
num_years = max_date.year - min_date.year + 1
#print(max_date.year)
#print(min_date.year)
print(num_years)
```
